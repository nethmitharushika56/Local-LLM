# Local LLM Notebook Demo

Use this notebook to exercise the Next.js chat API from Python. It posts a message to `/api/chat`, then prints the model answer and the retrieved context snippets.

## Prerequisites

Start the app in `my-app` before running the cells below:

```bash
npm run dev
```

The notebook assumes the app is available at `http://127.0.0.1:3000`.

````python
import json
from urllib.request import Request, urlopen
from urllib.error import HTTPError, URLError

BASE_URL = "http://127.0.0.1:3000"
CHAT_URL = f"{BASE_URL}/api/chat"

def post_json(url, payload):
    data = json.dumps(payload).encode("utf-8")
    request = Request(
        url,
        data=data,
        headers={"Content-Type": "application/json"},
        method="POST",
    )
    with urlopen(request, timeout=120) as response:
        return json.loads(response.read().decode("utf-8"))

print(f"Configured chat endpoint: {CHAT_URL}")
````

````python
message = "Explain how this app uses Ollama and ChromaDB."
history = [
    {"role": "user", "content": "What is this project?"},
    {"role": "assistant", "content": "It is a local-first chat app built with Next.js, Ollama, and ChromaDB."},
]

try:
    response = post_json(CHAT_URL, {"message": message, "history": history})
    print("Model:", response.get("model"))
    print("\nAnswer:\n")
    print(response.get("answer", ""))
    print("\nContext:\n")
    for index, item in enumerate(response.get("context", []), start=1):
        print(f"[{index}] {item.get('document', '')}")
        print(f"    metadata: {item.get('metadata', {})}")
        if item.get('distance') is not None:
            print(f"    distance: {item.get('distance')}")
        print()
except HTTPError as error:
    print("HTTP error:", error.code, error.reason)
    print(error.read().decode("utf-8"))
except URLError as error:
    print("Connection error:", error.reason)
````

## Notes

- The notebook uses the same API route as the web UI.
- If the request fails, check that Ollama is running on port `11434` and ChromaDB is running on port `8000`.
- You can change `message` and `history` in the code cell to test different prompts.
